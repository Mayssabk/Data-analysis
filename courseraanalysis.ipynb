```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
coursera_df=pd.read_csv(r"C:\Users\HP\Downloads\archive (1)\udemy_courses_dataset.csv")
sns.set(style="whitegrid")
#Histogram for number of subscribers
plt.figure(figsize=(10, 6))
plt.hist(coursera_df['num_subscribers'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Number of Subscribers')
plt.xlabel('Number of Subscribers')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar Chart for count of courses per subject
plt.figure(figsize=(10, 6))
coursera_df['subject'].value_counts().plot(kind='bar', color='lightgreen', edgecolor='black')
plt.title('Number of Courses per Subject')
plt.xlabel('Subject')
plt.ylabel('Number of Courses')
plt.show()
```

```python
# Scatter Plot for number of reviews vs. number of subscribers
plt.figure(figsize=(10, 6))
plt.scatter(coursera_df['num_reviews'], coursera_df['num_subscribers'], alpha=0.5, color='coral')
plt.title('Number of Reviews vs. Number of Subscribers')
plt.xlabel('Number of Reviews')
plt.ylabel('Number of Subscribers')
plt.show()
```

```python
# Heat Map of correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(coursera_df[['price', 'num_subscribers', 'num_reviews', 'num_lectures', 'content_duration']].corr(), 
            annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```
